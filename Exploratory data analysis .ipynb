# Exploratory data analysis form my thesis

## data preparation

```r
library(tidyverse)
library(reshape2)
library(cowplot)
```

```r
setwd("D:/")
dat <- read.csv("thes.csv")
dat$date <- as.Date(dat$date, format = "%d:%B:%Y")
dat$duration <- dat$end17-dat$onset17
```

## onset of emergence

```r
dat1<-dcast(dat %>% select(location,onset17,date) , date~location, value.var = "onset17")

t.test(dat1$Changkleua,dat1$Kham, paired =T)

p<- ggplot(dat, aes(x = location, y = onset17, color = location))+
  theme_classic2()

p1 <- p +  geom_boxplot()

p2 <- p + geom_jitter(position = position_jitter(width = 0.1, height = 0))

p3 <- ggplot(dat, aes(x = onset17, color = location, fill = location))+
  geom_density(alpha = 0.5)

p4 <- p +  geom_point(aes(group = date, color = factor(date)))+
  geom_line(aes(group = date, color = factor(date)))+
  theme(legend.position="none")

plot_grid(p1, p2, p3, p4, labels = c(LETTERS[1:4]), align = "v")
```

## end of emergence

```r
dat2<-dcast(dat %>% select(location,end17,date) , date~location, value.var = "end17")

t.test(dat2$Changkleua,dat1$Kham, paired =T)

p<- ggplot(dat, aes(x = location, y = end17, color = location))+
  theme_classic2()

p1 <- p +  geom_boxplot()

p2 <- p + geom_jitter(position = position_jitter(width = 0.1, height = 0))

p3 <- ggplot(dat, aes(x = end17, color = location, fill = location))+
  geom_density(alpha = 0.5)

p4 <- p +  geom_point(aes(group = date, color = factor(date)))+
  geom_line(aes(group = date, color = factor(date)))+
  theme(legend.position="none")

plot_grid(p1, p2, p3, p4, labels = c(LETTERS[1:4]), align = "v")
```
